```python
import pandas as pd
from geopy.geocoders import Nominatim
import plotly.express as px

geolocator = Nominatim(user_agent="cics397a")
```

```python
hosp_loc_df = pd.read_csv("hospital_locs.tsv", delimiter="\t").drop(["Website", "Map"], axis=1)
hosp_loc_df.set_index("Name", inplace=True)
hosp_loc_df
```

```python
hosp_size_df = pd.read_csv("hospital_sizes.tsv", delimiter="\t")
hosp_size_df.set_index("Hospital Name", inplace=True)

for col in ["Staffed Beds", "Total Discharges", "Patient Days"]:
    hosp_size_df[col] = hosp_size_df[col].str.replace(",", "").apply(int)

hosp_size_df
```

```python
hosp_df = pd.concat([hosp_size_df, hosp_loc_df], axis=1)  # Fix this line!
hosp_df
```

```python
hosp_df = hosp_df[hosp_df["Street Address"].notna()]
hosp_df = hosp_df[hosp_df["Staffed Beds"].notna()]
```

```python
hosp_df['address'] = hosp_df['Street Address'] + "," + " " + hosp_df['City'] + " MA, USA"
hosp_df
```

```python

hosp_df['Location'] = hosp_df['address'].apply(geolocator.geocode)
hosp_df
```

```python
hosp_df = hosp_df[hosp_df["Location"].notna()]
```

```python
hosp_df['Latitude'] = hosp_df['Location'].apply(lambda x: x.latitude)
hosp_df['Longitude'] = hosp_df['Location'].apply(lambda x: x.longitude)

hosp_df
```

```python
fig = px.scatter_mapbox(hosp_df, lat="Latitude", lon="Longitude", color="Longitude", size="Staffed Beds", zoom=7, size_max=20 )  # Fix this line!

fig.update_layout(mapbox_style="carto-positron")  # Fix this line!
```
